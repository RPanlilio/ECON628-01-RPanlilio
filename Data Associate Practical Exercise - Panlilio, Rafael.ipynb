## Cleaning Data Sets

```python
# Replace Missing Values
vip1 = vip1.replace(np.nan, '', regex=True)
vip2 = vip2.replace(np.nan, '', regex=True)
```

```python
# Change to Upper Case
vip2['Street'] = map(lambda x: x.upper(), vip2['Street'])
vip2['City'] = map(lambda x: x.upper(), vip2['City'])
```

```python
# Create Address Columns 
vip1['Address-Voter'] = vip1['Street'] + " " + vip1['City'] + " " +  vip1['State'] + " " + vip1['Zip'] + " " + vip1['Precinct ID']
vip2['Address-Polling'] = vip2['Street'] + " " + vip2['City'] + " " + vip2['State/ZIP'] + " " + vip2['Country'] + " " + vip2['Precinct']
```

```python
# Create Zip and State Columns
vip2['Zip'] =  vip2['State/ZIP'].str[3:]
vip2['State'] =  vip2['State/ZIP'].str[0:2]
```

```python
# Replace values in Precinct Polling Places Columns
vip2.loc[vip2.Precinct == '', 'Country'] = vip2['Address-Polling'].str.split().str[-2]
vip2.loc[vip2.Precinct == '', 'Zip'] = vip2['Address-Polling'].str.split().str[-3]
vip2.loc[vip2.Precinct == '', 'State'] = vip2['Address-Polling'].str.split().str[-4]
vip2.loc[vip2.Precinct == '', 'City'] = vip2['Address-Polling'].str.split().str[-5]
vip2.loc[vip2.Precinct == '', 'State/ZIP'] = vip2['State'] + " " + vip2['Zip']
```

```python
# Replace value in Precinct Column
vip2.loc[vip2.Precinct == '', 'Precinct'] = " " + vip2['Address-Polling'].str.split().str[-1]
```

```python
# Create Columns for Cleaning
vip2['Step1'] = " " + vip2['City'] + " " + vip2['State'] + " " + vip2['Zip'] + " " + vip2['Country'] + " " + vip2['Precinct']
vip2['Step2'] = vip2['Step1'].str.len()
```

```python
# List Comprehension to Extract Street String from Address String
for i in range(len(vip2)):
    vip2['Street'][i] = vip2['Address-Polling'][i][:-vip2['Step2'][i]]
```

```python
vip2
```

```python
# Create Column to Merge On
vip1['PrecinctMerge'] = vip1['Precinct ID'].str[4:]
vip2['PrecinctMerge'] = vip2['Precinct'].str[-3:]
```

## Before Merging the Data Sets, I needed to perform Data Cleaning. The Precinct Polling List Data Set had string data in columns it was not meant to be in. To fix this issue, my first step was to create a new column, Address-Polling, containing the complete address by putting all the string information in Street, City, State, Zip, Country, and Precinct columns together. My second step was to create a second column containing the string information in City, State, Zip, Country, and Precinct. I then performed List Comprehension replacing values in the Street column with the values of the Address-Polling column with the City, State, Zip, Country, and Precinct strings stripped from it. My last step was to create a column, PrecinctMerge, for which to merge the two data sets onto each other.   

# Merging Address and Precinct Poling Places Data Sets

```python
# Merge Data Sets
vip = pd.merge(vip1[['Address-Voter', 'City', 'State', 'Zip', 'Precinct ID', 'PrecinctMerge']], vip2[['Address-Polling', 'City', 'State', 'Zip', 'Precinct', 'PrecinctMerge']], how='left', on=['State', 'PrecinctMerge'])
```

```python
# Rename Columns
vip.rename(columns={'City_x':'City-Voter'}, inplace=True)
vip.rename(columns={'City_y':'City-Polling'}, inplace=True)
vip.rename(columns={'Zip_x':'Zip-Voter'}, inplace=True)
vip.rename(columns={'Zip_y':'Zip-Polling'}, inplace=True)
vip.rename(columns={'Precinct ID':'Precinct-Voter'}, inplace=True)
vip.rename(columns={'Precinct':'Precinct-Polling'}, inplace=True)
```

```python
# Reorder Columns
vip = vip[['Address-Voter', 'City-Voter', 'Zip-Voter', 'Precinct-Voter', 'Address-Polling', 'City-Polling', 'Zip-Polling', 'State', 'Precinct-Polling', 'PrecinctMerge']]
```

```python
vip = vip.fillna("No Matching Precinct Polling Place")
vip
```
